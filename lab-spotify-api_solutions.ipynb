```python
#function to search for songs
def search_song(title, artist):
    """
    Input
    title: song title
    artist: artist name
    limit: no. of results to return
    Output
    string with Spotify song id number
    """
    query = "tracks: "+title+" artist: "+artist
    #print(query)
    try:
        result = sp.search(query, limit=5)
        song_id = result['tracks']['items'][0]['id']
    except:
        song_id = "not_found"
    return song_id
```
